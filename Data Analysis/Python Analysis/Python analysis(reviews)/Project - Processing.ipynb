```python
import pandas as pd

df = pd.read_csv("Project_Data.csv")

df.head(20)
```

```python
df[["Reviewed Location","Date"]] = df["Date"].str.split(" on ",n = 1,expand = True)
df
```

```python
df["Reviewed Location"] = df["Reviewed Location"].str.strip("Reviewed in the ")
df 
```

```python
df[["Flavor Name","Size(Ounce)"]] = df["Style"].str.split("|",n = 1,expand = True)
df
```

```python
df["Flavor Name"] = df["Flavor Name"].str.strip("Flavor Name: ")
df["Size(Ounce)"] = df["Size(Ounce)"].str.strip("Size: ")
df["Size(Ounce)"] = df["Size(Ounce)"].str.strip("(Pack of 1)")
df["Size(Ounce)"] = df["Size(Ounce)"].str.strip(" Ounce")
df.head(20)
```

```python
df['Flavor Name'] = df["Flavor Name"].replace("Chees","Cheese")
df['Flavor Name'] = df["Flavor Name"].replace("Peanut Butt","Peanut Butter")
df
```

```python
df['Verified'] = df["Verified"].replace("-","No")
df['Helpful'] = df["Helpful"].replace("-","No")
df
```

```python
df = df.drop_duplicates()
df = df.drop(columns="Style")
df
```

```python
df.to_excel("Project_o2.xlsx")
```
