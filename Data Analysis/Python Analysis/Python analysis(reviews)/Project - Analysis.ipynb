Exploratory Data Analysis(EDA)
--------------

```python
verified_review_plot = sns.countplot(x = "Verified", data=df)

for bar in verified_review_plot.containers:
    verified_review_plot.bar_label(bar)
```

- Mostly Revivers were verified (91.3% were verified that is 10.5 times greater)

```python
helpful_review_plot = sns.countplot(x = "Helpful", data=df)

for bar in helpful_review_plot.containers:
    helpful_review_plot.bar_label(bar)
```

- Unfortunately, Almost 70% of the reviews were not helpful (2.5 times more than helpful reviews)

```python
location_plot = sns.countplot(x = "Reviewed Location", data=df)

for bar in location_plot.containers:
    location_plot.bar_label(bar)
```

- Almost all the reviews are from US(i.e. 98%) while only 12 are from canada and just 1 from UK

```python
rating_plot = sns.countplot(x = "Rating", data=df)

for bar in rating_plot.containers:
    rating_plot.bar_label(bar)
```

- Although most reviews were 5 star(i.e. 25.4%) but 1 star rewiew were also very high(i.e. 22.8%)
- 2, 3 and 4 star rating were very similar(i.e. 16% - 18%)

```python
flavor_plot = sns.countplot(x = "Flavor Name", data=df)

for bar in flavor_plot.containers:
    flavor_plot.bar_label(bar)
```

- Almost 89% of the reviews include cheese while only 11% include other flavors (i.e. 9 times more)

```python
rating_flavor_plot = sns.barplot(x = "Flavor Name", y = "Rating", data = df)

for bar in rating_flavor_plot.containers:
    rating_flavor_plot.bar_label(bar)
```

- Although Cheese is the most reviewed flavor but its average is 3-star rating as well as minmum and maximum are almost the same. Also there is no 4 or 5 star rating for cheese. 
- Maximum rating of Bacon, Peanut Butter and Chicken is almost the same that is more than 3.5 star rating which does reach to 5 star as well sa 4 star.These flavors average is approximately 3 stars but not quite 3 star.
