```python
import pandas as pd
from openpyxl import load_workbook
import re

def extract_age(string_age):
    age_match = re.search(r'\d+', string_age)
    if age_match:
        return int(age_match.group())
    else:
        return None

def get_age_group(age):
    if age is None:
        return 'Unknown'
    elif age < 18:
        return 'Under 18'
    elif 18 <= age < 30:
        return '18-29'
    elif 30 <= age < 45:
        return '30-44'
    elif 45 <= age < 60:
        return '45-59'
    else:
        return '60 and above'

file_path = 'FULLDATA.xlsx'

xls = pd.ExcelFile(file_path)

for sheet_name in xls.sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    
    # Apply the extract_age function to the 'Age' column
    df['AGE'] = df['AGE'].apply(extract_age)
    
    # Apply the get_age_group function to the extracted ages and create a new 'Age Group' column
    df['AGE_GROUP'] = df['AGE'].apply(get_age_group)
    
    with pd.ExcelWriter(file_path, engine='openpyxl', mode='a') as writer:
        writer.book = load_workbook(file_path)
        df.to_excel(writer, sheet_name=sheet_name, index=False)
```
