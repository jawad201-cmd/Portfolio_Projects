```python
import pandas as pd
import tensorflow as tf
```

```python
CSV_COL_NAME = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Species"]
SPECIES = ["Setosa", "Versicolor", "Virginica"]
```

```python
train_path = tf.keras.utils.get_file("iris_training.csv", "https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv")
test_path = tf.keras.utils.get_file("iris_testing.csv", "https://storage.googleapis.com/download.tensorflow.org/data/iris_test.csv")
```

```python
train = pd.read_csv(train_path, names=CSV_COL_NAME, header=0)
test = pd.read_csv(test_path, names=CSV_COL_NAME, header=0)

orignal_train = pd.read_csv(train_path, names=CSV_COL_NAME, header=0)
orignal_test = pd.read_csv(test_path, names=CSV_COL_NAME, header=0)
```

```python
train.head(10)
```

```python
train_y = train.pop("Species")
test_y = test.pop("Species")
```

```python
train = train.reset_index(drop=True)
```

```python
dnn_model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(train.shape[1],)),
    tf.keras.layers.Dense(64, activation="relu"),
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(3, activation="softmax")
])
```

```python
dnn_model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
training = dnn_model.fit(train, train_y, epochs=50, batch_size=256, validation_split=0.2)
```

```python
test_loss, test_acc = dnn_model.evaluate(test, test_y)
print("Test Accuracy: " , test_acc)
```

Predication Example:

```python
import numpy as np 
import sklearn as skl

sample = np.array([[3.1, 5.1, 4.0, 0.2]])
scaler = skl.preprocessing.StandardScaler()
scaler.fit(train)
sample_scaled = scaler.transform(sample)
prediction = dnn_model.predict(sample_scaled)
```
