```python
import pandas as pd
import tensorflow as tf
import tensorflow_probability as tfp
```

```python
tfd = tfp.distributions
inital_dist = tfd.Categorical(probs=[0.2, 0.8])
transition_dist = tfd.Categorical(probs=[[0.7, 0.3], [0.2, 0.8]])
observation_dist = tfd.Normal(loc=[0.0, 15.0], scale=[5.0, 10.0])
```

```python
hmm_model = tfd.HiddenMarkovModel(
    initial_distribution = inital_dist,
    transition_distribution = transition_dist,
    observation_distribution = observation_dist,
    num_steps = 7
)
```

```python
mean = hmm_model.mean()

with tf.compat.v1.Session() as sess:
    print(mean.numpy())
```
