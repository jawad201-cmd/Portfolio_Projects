Importing Packages & Cleaning:
=============================

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import sklearn as skl
import sklearn.feature_selection as sklf
import sklearn.model_selection as sklm
import sklearn.linear_model as skll
```

```python
df= pd.read_csv('Titanic-Dataset.csv')
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr
```

```python
df = df.drop("Cabin", axis = 1)
```

```python
df = df.dropna()
df.head(5)
```

```python
counts = df['Survived'].value_counts()
counts.plot.bar()
```

Modifing the Dataset for the Model:
----------------------------------

```python
age_groups = {
    '0-11': range(0, 11),
    '11-19': range(11, 19),
    '19-31': range(19, 31),
    '31-61': range(31, 61),
    '61-80': range(61, 80),
    '81+': range(80, 150),
}
def age_to_agegroup(age):
    for age_group, age_range in age_groups.items():
        if age in age_range:
            return age_group
    return 0

df['AgeGroup'] = df['Age'].apply(age_to_agegroup)

AgeGroup_df = pd.get_dummies(df['AgeGroup'])
AgeGroup_df
```

```python
gender_df = pd.get_dummies(df['Sex'])
gender_df
```

```python
Pclass_df = pd.get_dummies(df['Pclass'])
Pclass_df
```

```python
df["Fare"] = df["Fare"].astype(int)
fare_groups = {
    '0-50': range(0, 50),
    '50-100': range(50, 100),
    '100-150': range(100, 150),
    '150-200': range(150, 200),
    '200-300': range(200, 300),
    '300+': range(300, 1000),
}
def fare_to_faregroup(fare):
    for fare_group, fare_range in fare_groups.items():
        if fare in fare_range:
            return fare_group
    return 0

df['FareGroup'] = df['Fare'].apply(fare_to_faregroup)

FareGroup_df = pd.get_dummies(df['FareGroup'])
FareGroup_df
```

Predictive model(w.r.t Age):
===============================

```python
X= AgeGroup_df.iloc[:,:]
X.columns = X.columns.astype(str)
Y= df.iloc[:,1]
```

```python
#Using SelectKBest() and chi2 from sklearn.feature_selection/sklf:
best_features= sklf.SelectKBest(score_func=sklf.chi2, k=3)
fit= best_features.fit(X,Y)
```

```python
df_scores= pd.DataFrame(fit.scores_)
df_columns= pd.DataFrame(X.columns)
```

```python
features_scores= pd.concat([df_columns, df_scores], axis=1)
features_scores.columns= ['AgeGroup', 'Score']
features_scores.sort_values(by = 'Score')
```

```python
features_scores.plot.bar(x="AgeGroup")
```

```python
X= AgeGroup_df
X.columns = X.columns.astype(str)
Y= df[['Survived']]
Y = np.ravel(Y)
```

```python
#Using train_test_split() from sklearn.model_selection/sklm:
X_train, X_test, Y_train, Y_test = sklm.train_test_split(X, Y, test_size=0.4, random_state=100)
```

```python
#Using LogisticRegression() from sklearn.linear_model/skll:
logreg= skll.LogisticRegression()
logreg.fit(X_train,Y_train)
```

```python
Y_pred=logreg.predict(X_test)
print (X_test) #test dataset
print (Y_pred) #predicted values
```

Precision & ROC Curve:
----------------------

```python
#Using metrics from sklearn/skl:

print("Accuracy: ",skl.metrics.accuracy_score(Y_test, Y_pred))
print("Recall: ",skl.metrics.recall_score(Y_test, Y_pred, zero_division=1))
print("Precision:",skl.metrics.precision_score(Y_test, Y_pred, zero_division=1))
print("CL Report:",skl.metrics.classification_report(Y_test, Y_pred, zero_division=1))
```

```python
y_pred_proba= logreg.predict_proba(X_test) [::,1]
false_positive_rate, true_positive_rate, _ = skl.metrics.roc_curve(Y_test, y_pred_proba)
auc= skl.metrics.roc_auc_score(Y_test, y_pred_proba)

plt.plot(false_positive_rate, true_positive_rate,label="AUC="+str(auc))
plt.title('ROC Curve')
plt.ylabel('True Positive Rate')
plt.xlabel('false Positive Rate')
plt.legend(loc=4)
```
