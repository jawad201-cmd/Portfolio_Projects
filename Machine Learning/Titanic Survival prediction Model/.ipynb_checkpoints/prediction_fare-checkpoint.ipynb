```python
df = df.drop("Cabin", axis = 1)
```

```python
df = df.dropna()
df.head(5)
```

```python
df["Fare"] = df["Fare"].astype(int)
fare_groups = {
    '0-50': range(0, 50),
    '50-100': range(50, 100),
    '100-150': range(100, 150),
    '150-200': range(150, 200),
    '200-300': range(200, 300),
    '300+': range(300, 1000),
}
def fare_to_faregroup(fare):
    for fare_group, fare_range in fare_groups.items():
        if fare in fare_range:
            return fare_group
    return 0

# Apply the function to create a new column 'AgeGroup'
df['FareGroup'] = df['Fare'].apply(fare_to_faregroup)
df.head(10)
```

```python
FareGroup_df = pd.get_dummies(df['FareGroup'])
#AgeGroup_df = AgeGroup_df.astype(int)
FareGroup_df
```

```python
X= FareGroup_df.iloc[:,:]
#X.columns = X.columns.astype(str)
Y= df.iloc[:,1]
```

Predictive model(w.r.t Fare):
===============================

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

best_features= SelectKBest(score_func=chi2, k=3)
fit= best_features.fit(X,Y)
```

```python
df_scores= pd.DataFrame(fit.scores_)
df_columns= pd.DataFrame(X.columns)
```

```python
features_scores= pd.concat([df_columns, df_scores], axis=1)
features_scores.columns= ['AgeGroup', 'Score']
features_scores.sort_values(by = 'Score')
```

```python
X= FareGroup_df
X.columns = X.columns.astype(str)
Y= df[['Survived']]
Y = np.ravel(Y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression

logreg= LogisticRegression()
logreg.fit(X_train,Y_train)
```

```python
Y_pred=logreg.predict(X_test)
print (X_test) #test dataset
print (Y_pred) #predicted values
```

Precision & ROC Curve:
=====================

```python
from sklearn import metrics
from sklearn.metrics import classification_report

print("Accuracy: ",metrics.accuracy_score(Y_test, Y_pred))
print("Recall: ",metrics.recall_score(Y_test, Y_pred, zero_division=1))
print("Precision:",metrics.precision_score(Y_test, Y_pred, zero_division=1))
print("CL Report:",metrics.classification_report(Y_test, Y_pred, zero_division=1))
```

```python
y_pred_proba= logreg.predict_proba(X_test) [::,1]
```

```python
false_positive_rate, true_positive_rate, _ = metrics.roc_curve(Y_test, y_pred_proba)
```

```python
auc= metrics.roc_auc_score(Y_test, y_pred_proba)
```

```python
plt.plot(false_positive_rate, true_positive_rate,label="AUC="+str(auc))
plt.title('ROC Curve')
plt.ylabel('True Positive Rate')
plt.xlabel('false Positive Rate')
plt.legend(loc=4)
```
